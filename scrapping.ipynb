```python
#Library untuk melakukan Scrapping
from google_play_scraper import app, reviews_all, Sort, reviews
import pandas as pd
```

#Scrapping Data Playstore

```python
# ID aplikasi yang akan di-scrape ulasannya
app_id = 'app.source.getcontact'
```

```python
# Mengambil  sebagian ulasan get contact
scrapping = reviews(
    app_id,
    lang='id', # Bahasa Indonesia
    country='id', # Negara Indonesia
    sort=Sort.MOST_RELEVANT,
    count = 15000, # 15 ribu ulasan yang diambil
)
```

```python
# Menyimpan ulasan ke file CSV

data = pd.DataFrame(scrapping[0])
data.to_csv('getcontact.csv', index=False)
```
